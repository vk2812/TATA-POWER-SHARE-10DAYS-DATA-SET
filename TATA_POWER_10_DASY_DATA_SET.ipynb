```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tatapower=pd.read_excel('TATA POWER 10D.xlsx')
tatapower.head()
```

```python
tatapower.shape
```

```python
tatapower.isnull
```

```python
sns.set_style('dark')
sns.countplot(x='Low',data=tatapower)
```

```python
sns.set_style('dark')
sns.countplot(x='Close',data=tatapower)
```

```python
tatapower['Volume'].hist(color='red', bins=20, figsize=(8,4))
```

```python

sns.countplot(tatapower['Close'])
plt.title('previous close', fontsize=5)
plt.xlabel('Date')
plt.ylabel('Open')
plt.show()
```

```python
tatapower.info()
```

```python
tatapower.drop_duplicates(inplace = True)
```

```python
tatapower.corr()
#Note: The corr() method ignores "not numeric" columns.
```

```python
sns.countplot(x='Date',data=tatapower)
```

```python
sns.kdeplot(x = 'Date' , data = tatapower , color = 'black')
```

```python
sns.kdeplot(x = 'Volume' , data = tatapower , color = 'red')
```

```python
sns.displot(x = 'Volume',kde=True,bins = 5 , data =tatapower)
```

```python
sns.displot(x = 'Close', kde=True,bins = 5 , data =tatapower)
sns.displot(y = 'Date', kde=True,bins = 5 , data =tatapower)
#kde - It is set to False by default. However, if you wish to plot a KDE graph on top of the bars, you can set it to True.

#bins - The number of bins/bars. The lower the number, wider the bars and wider the intervals.
```

```python
sns.displot(x = 'Close', kde=False,bins = 5 , data =tatapower)
```

```python
sns.scatterplot(x='High', y ='Date' ,
data = tatapower , hue = 'Volume')
```

```python
corr = tatapower.corr()
sns.heatmap(corr)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x = 'High' , data = tatapower,
hue = 'Date')
```

```python
sns.jointplot(x = 'Open', y = 'Close',
data = tatapower,hue = 'Date')
```
